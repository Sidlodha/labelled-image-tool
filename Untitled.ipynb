```python
import requests
```

```python
for i in range(250,501):   
    url = "https://uplara.herokuapp.com/v1/graphql"
    json = {"operationName":"MyQuery3",
            "variables":{"product_id": str(i)},
            "query":"query MyQuery3($product_id: Int!) {\n  product_images(where: {product_id: {_eq: $product_id}}) {  isLabelled\n category\n item_category\n}\n}\n"}
    headers = {
        "x-hasura-admin-secret": "Labeling93*"
    }
    x = requests.post(url = url, json = json, headers = headers)
    if x.status_code == 200:
        data = x.json()['data']
        clothing = False
        person = False
        item_category = None
        product = data["product_images"]
        if product:
            for eachImage in product:
                if eachImage["isLabelled"]=="Labelled" and eachImage["category"]=="Person":
                    person = True
                if eachImage["isLabelled"]=="Labelled" and eachImage["category"]=="Clothing":
                    clothing = True
                    item_category = eachImage["item_category"]
            if clothing == True and person == True:
                condition_json = {
                    "operationName": "MyMutation2",
                    "variables":{"product_id":str(i),"condition":"good", "item_category": item_category},
                    "query":"mutation MyMutation2($product_id: Int!, $condition: String!, $item_category: String!) {\n  update_product_condition(where: {product_id: {_eq: $product_id}}, _set: {condition: $condition, item_category: $item_category}) {\n    affected_rows\n    __typename\n  }\n}\n"
                }
                conf = requests.post(url = url, json = condition_json, headers = headers)
                if conf.status_code == 200:
                    conf_data = conf.json()['data']
                    print(conf_data,"good",i)
            else:
                condition_json = {
                    "operationName": "MyMutation2",
                    "variables":{"product_id":str(i),"condition":"bad"},
                    "query":"mutation MyMutation2($product_id: Int!, $condition: String!) {\n  update_product_condition(where: {product_id: {_eq: $product_id}}, _set: {condition: $condition}) {\n    affected_rows\n    __typename\n  }\n}\n"
                }
                conf = requests.post(url = url, json = condition_json, headers = headers)
                if conf.status_code == 200:
                    conf_data = conf.json()['data']
                    print(conf_data,"bad",i)
    
```
